```python
import os
import cv2
import face_recognition
import numpy as np
from datetime import datetime
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def create_folder_for_images():
    now = datetime.now()
    folder_name = now.strftime('%Y%m%d_%H%M%S')
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    return folder_name

def augment_image(image):
    datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        brightness_range=[0.8, 1.2],
        channel_shift_range=0.1,
        fill_mode='nearest'
    )
    
    # 이미지를 rank 4로 변환 (채널 축 추가)
    image = image.reshape((1,) + image.shape + (1,))
    aug_iter = datagen.flow(image, batch_size=1)
    aug_images = [next(aug_iter)[0].astype(np.uint8) for _ in range(10)]  # 각 이미지당 10개의 증강 이미지 생성
    
    # 증강된 이미지에서 채널 축 제거
    aug_images = [aug_img.reshape(aug_img.shape[:-1]) for aug_img in aug_images]
    return aug_images

def capture_and_augment_images(folder_name, num_images=100):  # 사진 찍는 횟수
    video_capture = cv2.VideoCapture(0)
    count = 0

    try:
        while count < num_images:
            ret, frame = video_capture.read()
            if not ret:
                break

            rgb_frame = frame[:, :, ::-1]
            face_locations = face_recognition.face_locations(rgb_frame)

            if face_locations:
                count += 1
                image_path = os.path.join(folder_name, f'image_{count}.jpg')
                cv2.imwrite(image_path, frame)
                print(f'Captured {count}/{num_images}')

                gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                augmented_images = augment_image(gray_frame)
                for i, aug_img in enumerate(augmented_images):
                    aug_img_bgr = cv2.cvtColor(aug_img, cv2.COLOR_GRAY2BGR)
                    aug_image_path = os.path.join(folder_name, f'image_{count}_aug_{i+1}.jpg')
                    cv2.imwrite(aug_image_path, aug_img_bgr)

            cv2.imshow('Video', frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    except Exception as e:
        print(f"Error during capturing or augmenting images: {e}")
    finally:
        video_capture.release()
        cv2.destroyAllWindows()

def main_capture():
    folder_name = create_folder_for_images()
    capture_and_augment_images(folder_name)
    
    new_name = input("폴더 이름을 사용자의 이름으로 변경하세요: ")
    os.rename(folder_name, new_name)
    print(f'폴더 이름이 {new_name}(으)로 변경되었습니다.')

if __name__ == "__main__":
    main_capture()
```

```python
import os
import pickle
import face_recognition
from sklearn import svm
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier

def load_faces_from_dir(directory):
    X = []
    y = []
    supported_formats = ['.jpg', '.jpeg', '.png']

    for root, dirs, files in os.walk(directory):
        for filename in files:
            if any(filename.lower().endswith(ext) for ext in supported_formats):
                path = os.path.join(root, filename)
                image = face_recognition.load_image_file(path)
                face_encodings = face_recognition.face_encodings(image)
                if face_encodings:
                    X.append(face_encodings[0])
                    label = os.path.basename(root)
                    y.append(label)
                    print(f"Loaded {filename} from {label}")
                else:
                    print(f"No faces found in {filename}")
            else:
                print(f"Skipped {filename}, unsupported file type")
    if not X:
        print("No face encodings were loaded. Check your dataset and file types.")
    return X, y

def train_ensemble(known_faces_dir, model_save_path='ensemble_model2.pkl'):
    print("Loading faces from directory...")
    X, y = load_faces_from_dir(known_faces_dir)
    print("Faces loaded. Total samples:", len(X))

    print("Training SVM model...")
    svm_clf = svm.SVC(C=1, gamma=0.01, kernel='rbf', probability=True)
    svm_clf.fit(X, y)
    print("SVM model trained.")

    print("Training KNN model...")
    knn_clf = KNeighborsClassifier(n_neighbors=3)
    knn_clf.fit(X, y)
    print("KNN model trained.")

    print("Training RandomForest model...")
    rf_clf = RandomForestClassifier(n_estimators=100)
    rf_clf.fit(X, y)
    print("RandomForest model trained.")

    print("Training AdaBoost model...")
    adb_clf = AdaBoostClassifier(n_estimators=50)
    adb_clf.fit(X, y)
    print("AdaBoost model trained.")

    print("Training ensemble model...")
    ensemble_clf = VotingClassifier(estimators=[
        ('svm', svm_clf),
        ('knn', knn_clf),
        ('rf', rf_clf),
        ('adb', adb_clf)
    ], voting='soft')
    ensemble_clf.fit(X, y)
    print("Ensemble model trained.")

    with open(model_save_path, 'wb') as f:
        pickle.dump(ensemble_clf, f)

    print(f"앙상블 모델이 '{model_save_path}'에 저장되었습니다.")
    
train_ensemble(r'C:\Users\djjin\출퇴근관리시스템')
```

```python
import cv2
import mediapipe as mp
import face_recognition
import numpy as np
import pickle
from scipy.spatial import distance
import time
import os
from PIL import ImageFont, ImageDraw, Image
import warnings

# 시스템 폰트 경로 설정 (Windows 예시)
if os.name == 'nt':  # Windows
    font_path = "C:\\Windows\\Fonts\\malgun.ttf"  # 맑은 고딕 폰트 경로
else:
    # 기타 운영체제의 경우 적절한 폰트를 설정
    font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"

# MediaPipe Hands 모델 로드
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# 허공에 가상 버튼 위치 정의 (x축 변경, y축 고정)
button_positions = {
    '출근': (150, 600),
    '외출': (450, 600),
    '복귀': (750, 600),
    '퇴근': (1050, 600),
}

# 제스처 지속 시간 설정 (초)
gesture_hold_time = 2

# 가상 버튼 글씨 크기와 굵기 설정
button_font_size = 50
button_font = ImageFont.truetype(font_path, button_font_size)

# 문구 글씨 크기와 굵기 설정
text_font_size = 100
text_font = ImageFont.truetype(font_path, text_font_size)

# mediapipe 얼굴 검출 및 랜드마크 모델 로드
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

# 눈 깜빡임과 입 움직임 감지를 위한 임계값 설정
EYE_AR_THRESH = 0.2
MAR_THRESH = 0.5
STATIC_FRAME_COUNT_THRESH = 15  # 얼굴이 움직이지 않는 프레임 수 임계값
AUTHENTICATION_TIME = 5  # 인증 시간 (초)

# 눈 깜빡임 비율 계산 함수
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear

# 입 벌림 비율 계산 함수
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[2], mouth[10])  # 51, 59
    B = distance.euclidean(mouth[4], mouth[8])   # 53, 57
    C = distance.euclidean(mouth[0], mouth[6])   # 49, 55
    mar = (A + B) / (2.0 * C)
    return mar

# 가상 버튼 그리기 함수
def draw_buttons(image, positions, font):
    # OpenCV 이미지를 PIL 이미지로 변환
    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(pil_image)
    
    # 각 버튼 위치에 텍스트와 버튼 그리기
    for text, pos in positions.items():
        text_bbox = draw.textbbox((0, 0), text, font=font)
        text_width = text_bbox[2] - text_bbox[0]
        text_height = text_bbox[3] - text_bbox[1]
        text_x = pos[0] - text_width // 2
        text_y = pos[1] - text_height // 2
        draw.rectangle([(pos[0] - 50, pos[1] - 50), (pos[0] + 50, pos[1] + 50)], outline="white", width=2)
        draw.text((text_x, text_y), text, font=font, fill="black")
    
    # PIL 이미지를 다시 OpenCV 이미지로 변환
    return cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2BGR)

# 텍스트 그리기 함수
def draw_text(image, text, position, font):
    # OpenCV 이미지를 PIL 이미지로 변환
    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(pil_image)
    draw.text(position, text, font=font, fill="yellow")
    # PIL 이미지를 다시 OpenCV 이미지로 변환
    return cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2BGR)

# 손 제스처 감지 함수
def detect_gesture(image, results, positions):
    # 손 랜드마크를 기반으로 가상 버튼 위치와 손 위치 비교
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for id, lm in enumerate(hand_landmarks.landmark):
                h, w, c = image.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                for action, pos in positions.items():
                    if abs(cx - pos[0]) < 50 and abs(cy - pos[1]) < 50:
                        return action
    return None

# 액션 로그 기록 함수
def log_action(user, action):
    # 출근/퇴근/외출/복귀 액션 로그를 파일에 기록
    with open('attendance_log.txt', 'a', encoding='utf-8') as f:
        f.write(f"{time.strftime('%Y-%m-%d %H:%M:%S')}, {user}, {action}\n")

# 얼굴 인식 및 제스처 인식을 통해 사용자 인증 함수
def recognize_faces_and_gestures(model_path='ensemble_model.pkl'):
    # 앙상블 모델 로드
    with open(model_path, 'rb') as f:
        ensemble_clf = pickle.load(f)

    video_capture = cv2.VideoCapture(0)  # 웹캠 초기화
    video_capture.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)  # 해상도 설정
    video_capture.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)  # 해상도 설정
    previous_face_encodings = None
    static_frame_count = 0
    authentication_start_time = time.time()
    authenticated = False
    recognized_user = None
    recognized_prob = None

    gesture_start_time = None
    current_action = None
    action_confirmed = False

    # 얼굴 검출 모델 초기화
    with mp_face_detection.FaceDetection(min_detection_confidence=0.5) as face_detection:
        while True:
            ret, frame = video_capture.read()
            if not ret:
                print("카메라에서 프레임을 가져오지 못했습니다.")
                break

            # 프레임을 RGB로 변환
            rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = face_detection.process(rgb_frame)

            face_locations = []
            face_encodings = []

            # 얼굴 검출 결과 처리
            if results.detections:
                for detection in results.detections:
                    bboxC = detection.location_data.relative_bounding_box
                    ih, iw, _ = frame.shape
                    (startX, startY, endX, endY) = (int(bboxC.xmin * iw), int(bboxC.ymin * ih),
                                                    int((bboxC.xmin + bboxC.width) * iw), int((bboxC.ymin + bboxC.height) * ih))
                    face_locations.append((startY, endX, endY, startX))
                    encodings = face_recognition.face_encodings(rgb_frame, [(startY, endX, endY, startX)])
                    if encodings:
                        face_encodings.append(encodings[0])

            face_movement_detected = False
            if face_encodings:
                for face_encoding in face_encodings:
                    landmarks = face_recognition.face_landmarks(rgb_frame, face_locations)[0]
                    leftEye = np.array(landmarks["left_eye"])
                    rightEye = np.array(landmarks["right_eye"])
                    leftEAR = eye_aspect_ratio(leftEye)
                    rightEAR = eye_aspect_ratio(rightEye)
                    ear = (leftEAR + rightEAR) / 2.0

                    mouth = np.array(landmarks["top_lip"] + landmarks["bottom_lip"])
                    mar = mouth_aspect_ratio(mouth)

                    # 눈 깜빡임 또는 입 벌림 감지
                    if ear < EYE_AR_THRESH or mar > MAR_THRESH:
                        face_movement_detected = True

            if previous_face_encodings is not None:
                if face_encodings:
                    for prev_enc, enc in zip(previous_face_encodings, face_encodings):
                        if not np.allclose(prev_enc, enc, atol=0.02):
                            static_frame_count = 0
                            break
                    else:
                        static_frame_count += 1

                if static_frame_count >= STATIC_FRAME_COUNT_THRESH:
                    static_frame_count = 0
            previous_face_encodings = face_encodings

            if face_encodings and static_frame_count < STATIC_FRAME_COUNT_THRESH:
                if face_movement_detected:
                    if time.time() - authentication_start_time <= AUTHENTICATION_TIME:
                        predictions = ensemble_clf.predict(face_encodings)
                        probabilities = ensemble_clf.predict_proba(face_encodings)
                        recognized_names = [name if max(prob) >= 0.9 else "Unknown" for name, prob in zip(predictions, probabilities)]
                        recognized_probs = [max(prob) for prob in probabilities]

                        if "Unknown" not in recognized_names:
                            authenticated = True
                            recognized_user = recognized_names[0]
                            recognized_prob = recognized_probs[0]
                            print(f"Authenticated as: {recognized_user} with probability: {recognized_prob:.2f}")
                            break

            for (top, right, bottom, left) in face_locations:
                # 얼굴 영역에 사각형 그리기
                cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
                if recognized_user:
                    # 인식된 사용자의 이름 표시
                    cv2.putText(frame, recognized_user, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)
                else:
                    # 인식되지 않은 경우 'Unknown' 표시
                    cv2.putText(frame, "Unknown", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

            # 비디오 출력
            cv2.imshow('Video', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

            # 인증 시간 초과 시 루프 종료
            if time.time() - authentication_start_time > AUTHENTICATION_TIME:
                break

    video_capture.release()
    cv2.destroyAllWindows()

    return authenticated, recognized_user

# 얼굴 인식을 통해 사용자 인증
authenticated, recognized_user = recognize_faces_and_gestures()

# 인증된 사용자가 "Unknown"이 아닌 경우에만 제스처 인식 실행
if authenticated and recognized_user != "Unknown":
    print(f"User authenticated as {recognized_user}. Proceed with gesture recognition.")
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

    gesture_start_time = None
    current_action = None
    action_confirmed = False

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # 이미지 반전 및 전처리
        frame = cv2.flip(frame, 1)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # MediaPipe Hands로 손 제스처 인식
        results = hands.process(frame_rgb)
        
        # 버튼 그리기
        frame = draw_buttons(frame, button_positions, button_font)
        
        # 제스처 감지
        action = detect_gesture(frame, results, button_positions)
        
        if action and authenticated:
            if current_action == action:
                if time.time() - gesture_start_time >= gesture_hold_time:
                    frame = draw_buttons(frame, button_positions, button_font)
                    frame = draw_text(frame, f'{action}을 완료하였습니다.', (50, 50), text_font)
                    action_confirmed = True
                    log_action(recognized_user, action)
                    break  # 제스처 완료 후 종료
            else:
                current_action = action
                gesture_start_time = time.time()
                action_confirmed = False
        else:
            current_action = None
            gesture_start_time = None
            action_confirmed = False
        
        if current_action and not action_confirmed:
            frame = draw_buttons(frame, button_positions, button_font)
            frame = draw_text(frame, f'{current_action}하시겠습니까?', (50, 50), text_font)
        
        # 화면 출력
        cv2.imshow('Virtual Button Interaction', frame)
        
        # 'q' 키를 누르면 종료
        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
else:
    print("User authentication failed. The system will now exit.")
```
